### КРОК 6. РОЗГАЛУЖЕННЯ

```python

# ПРИКЛАДИ РОЗГАЛУЖЕНЬ В PYTHON

from random import randint

#  Приклад 1: тернарний if-else з випадковим числом ---
A = randint(0, 1)
print(f"Приклад 1 (тернарний if-else): {'Значить A='+str(A) if A else 'Але може бути що A='+str(A)}")

#  Приклад 2: простий if-else ---
x = 10
if x > 5:
    print("Приклад 2 (простий if-else): x більше 5")
else:
    print("Приклад 2 (простий if-else): x менше або рівне 5")

#  Приклад 3: if-elif-else ---
score = 75
if score >= 90:
    print("Приклад 3 (if-elif-else): Відмінно!")
elif score >= 60:
    print("Приклад 3 (if-elif-else): Добре, але можна краще")
else:
    print("Приклад 3 (if-elif-else): Потрібно ще попрацювати")
```

### КРОК 7. ОБРОБКА ПОМИЛОК

```python
#Конструкція try->except->finally
x = "10"
y = 5

try:
    # Намагання скласти рядок і число → помилка TypeError
    result = x + y
    print("Результат:", result)
except TypeError as e:
    print("Ой! Виникла помилка типу:", e)
finally:
    print("Цей блок finally виконується завжди")
```

### КРОК 8. КОНТЕКСТ-МЕНЕДЖЕР

```python
#Контекст-менеджер with

# Запис тексту у файл README.md
with open("README.md", "w") as f:
    f.write("Привіт! Це мій приклад файлу README.md.\n")
    f.write("Другий рядок тексту для демонстрації контекст-менеджера.\n")

# Зчитування файлу README.md
with open("README.md", "r") as f:
    for idx, line in enumerate(f):
        print(f"{idx})> {line.strip()}")
```

### КРОК 9. ЛЯМБДА-ФУНКЦІЇ

```python
# Список чисел
numbers = [1, 2, 3, 4, 5]

# Лямбда-функція для піднесення числа до квадрату
square = lambda x: x ** 2

# Використовуємо map, щоб застосувати лямбду до всіх елементів списку
squared_numbers = list(map(square, numbers))

print("Оригінальний список:", numbers)
print("Квадрати чисел:", squared_numbers)
```

```python
# ...existing code...

# ======= Основи: змінні та типи =========
a = "Привіт"           # str
b = 42                 # int
c = 3.14               # float
d = [1, 2, 3]          # list
e = {"x": 1, "y": 2}   # dict

print("Змінні і їх типи:")
print("a:", a, type(a))
print("b:", b, type(b))
print("c:", c, type(c))
print("d:", d, type(d))
print("e:", e, type(e))
print("-" * 40)

# ======= Арифметика та модуль math =========
import math
radius = 5
area = math.pi * radius ** 2
print(f"Площа круга радіуса {radius} = {area:.2f}")  # формат з 2 знаками
print("-" * 40)

# ======= Робота з рядками =========
s = "Python в Jupyter"
print("Довжина рядка:", len(s))
print("Слайсинг:", s[0:6])
print("Заміна:", s.replace("Jupyter", "Notebook"))
print("Розбиття:", s.split())
print("-" * 40)

# ======= Списки та list comprehension =========
nums = list(range(1, 11))
squares = [x**2 for x in nums]          # comprehension
evens = [x for x in nums if x % 2 == 0] # filter в comprehension
print("nums:", nums)
print("squares:", squares)
print("evens:", evens)
print("-" * 40)

# ======= Словники (dict) =========
person = {"name": "Khrystyna", "city": "Lviv", "age": 20}
print("Ім'я:", person.get("name"))
for k, v in person.items():
    print(f"{k} -> {v}")
print("-" * 40)

# ======= Функції =========
def greet(name: str) -> str:
    """Повертає привітальний рядок."""
    return f"Вітаю, {name}!"

print(greet("Світ"))
print("-" * 40)

# ======= Рекурсія / ітерація: факторіал =========
def factorial(n: int) -> int:
    """Обчислення факторіалу ітераційно (без переповнення стеку)."""
    if n < 0:
        raise ValueError("n має бути невід'ємним")
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

print("5! =", factorial(5))
print("-" * 40)

# ======= Lambda, map, filter =========
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x*x, nums))
filtered = list(filter(lambda x: x % 2 == 1, nums))
print("squared via map:", squared)
print("odd numbers via filter:", filtered)
print("-" * 40)

# ======= Обробка помилок (exceptions) =========
def safe_div(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Ділення на нуль!"
    except TypeError:
        return "Невірні типи для ділення"

print("10/2 =", safe_div(10, 2))
print("10/0 =", safe_div(10, 0))
print("10/'a' =", safe_div(10, "a"))
print("-" * 40)

# ======= Корисна порада для Jupyter =========
# Кожен блок краще зберігати в окремому cell для зручності відладки.
# Виконуй cell по черзі, читай вивід, змінюй змінні й повторно запускай.
# ...existing code...
```
