```python
class Item:
    """Це наш клас який описує ітем в грі"""
    def __init__(self):
        # тепер для кожного обєкта є характерним дані атрибути
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])
        self.rarity = "green"
        self.attack_power = random.randint(4, 6)
        self.durability = random.randint(30, 45)
        self.magic_attribute = "Магія Природи"
        self.description = "Рандомний предмет зброї"
    
    def attack(self):
        if "Магія" in self.magic_attribute:
            attack = self.attack_power + 2
        return f"{self.name} наніс {attack} шкоди типом {self.magic_attribute}"
    def defence(self):
        self.durability =+ 1
        return f"{self.name} захищається та відновлює міцність до {self.durability}"
        

for i in range(10):
    item = Item()
    print(f"""
            Подія {i}: Нам випав ітем:
              {item.name} з параметрами:
              Атака/захист: {item.attack_power}/{item.durability}
              Магічними властивостями: {item.magic_attribute}
            """)
    if random.randint(0,1):
        print(f"Застосовуємо предмет: {item.attack()}")
    else:
        print(f"Застосовуємо предмет: {item.defence()}")
```

```python
class Item:
    """Це наш клас який описує ітем в грі"""
    def __init__(self):
        # тепер для кожного обєкта є характерним дані атрибути
        self._initial_durability = random.randint(30, 45)
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])
        self.rarity = "green"
        self.attack_power = random.randint(4, 6)
        self.durability = self._initial_durability
        self.magic_attribute = "Магія Природи"
        self.description = "Рандомний предмет зброї"
    
    @property
    def get_attack_power(self):
        return self.attack_power
    
    @property
    def get_current_durability(self):
        return f"{self.name} має міцність {self.durability}/{self._initial_durability}"
```

```python
card = Item()
print(f"{card.attack_power} = {card.get_attack_power}")
card.attack_power += 1
print(card.attack_power, card.get_attack_power)
# Це буде помилка
#card.get_attack_power = 5
```

```python
import random
class Item:
    global_item_count = 0  # класова змінна
    """Це наш клас який описує ітем в грі"""
    def __init__(self):
        # тепер для кожного обєкта є характерним дані атрибути
        self._initial_durability = random.randint(30, 45)
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])

    def test_method(self):
        Item.global_item_count += 1
        return f"{self.name} має міцність {self._initial_durability}"
    
    @staticmethod
    def always_say_hello():
        return "Hello from Item class!"
    
i = Item
b = Item()
print(type(i), type(b))
# i.test_method() #TypeError: Item.test_method() missing 1 required positional argument: 'self'

print(b.test_method(), i.test_method(b))
print(b.always_say_hello(), i.always_say_hello())
```

```python
a = Item()
b = Item()
print(a.name, b.name)
print(a.test_method())
print("Зверху міняємо обєкт a а зміни можумо бачити в обєкті b:", b.global_item_count)

Item.global_item_count = 0
print("Зверху міняємо обєкт a а зміни можумо бачити в обєкті b:", Item.global_item_count)
print(b.test_method())
print("Зверху міняємо обєкт a b зміни можумо бачити в обєкті ab:", a.global_item_count)

c = Item()
print("Зверху міняємо обєкт a b зміни можумо бачити в обєкті ab:", c.global_item_count)
```

```python
class MyName:
    total_names = 0  # Змінна класу (підрахунок усіх об’єктів)

    def __init__(self, name=None):
        # Якщо ім'я не вказано → створюємо Anonymous
        if name is None:
            name = "Anonymous"

        # Якщо є цифри або символи → помилка
        if not name.isalpha():
            raise ValueError("Ім'я може містити лише літери!")

        # Робимо першу букву великою
        self.name = name.capitalize()

        # Збільшуємо кількість створених об’єктів
        MyName.total_names += 1
        self.my_id = MyName.total_names

    # Метод для створення пошти з можливістю змінити домен
    def create_email(self, domain="itcollege.lviv.ua"):
        return f"{self.name.lower()}@{domain}"

    # Метод, який повертає привітання (можна змінити текст)
    @staticmethod
    def say_hello(message="Hello to everyone!"):
        return f"You say: {message}"

    # Метод для підрахунку кількості букв у імені
    def name_length(self):
        return len(self.name)

    # Властивість, що повертає повну інформацію про користувача
    @property
    def full_name(self):
        return f"User #{self.my_id}: {self.name} ({self.create_email()})"

    # Метод, який зберігає користувача у файл
    def save_to_file(self, filename="users.txt"):
        with open(filename, "a", encoding="utf-8") as f:
            f.write(self.full_name + "\n")


# --- Перевірка роботи ---
print("Розпочинаємо створювати об’єкти!\n")

names = ("Bohdan", "Marta", "Khrystyna", None)
all_names = [MyName(name) for name in names]

for person in all_names:
    print(person.full_name)
    print("Кількість букв у імені:", person.name_length())
    print(person.say_hello(f"Привіт, я {person.name}!"))
    person.save_to_file()

print(f"\nМи створили {MyName.total_names} користувачів!")
```
