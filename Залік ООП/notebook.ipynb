Вимоги до класу:

- у програмі необхідно створити 2–3 об'єкти класу з різними вхідними даними та продемонструвати роботу класу.



AI Prompt

Створи клас CommonCharactersCounter мовою Python.
Клас має приймати два рядки, підраховувати кількість однакових символів,
містити метод обробки даних, property для отримання результату
та метод str для виведення інформації про студента.
Усі коментарі та вивід мають бути українською мовою.

```python
class CommonCharactersCounter:
    def __init__(self, string1: str, string2: str):
        # Зберігаємо вхідні рядки
        self.string1 = string1
        self.string2 = string2
        self._common_count = None

    def count_common(self):
        # Підраховуємо кількість однакових символів
        count = 0
        for char in self.string1:
            if char in self.string2:
                count += 1
        self._common_count = count

    @property
    def common_count(self):
        # Повертаємо кількість однакових символів
        if self._common_count is None:
            self.count_common()
        return self._common_count

    def __str__(self):
        # Вивід інформації про студента
        return "Ім'я студента: Богдан, Група: КН-31"


# Створення декількох об'єктів класу з різними вхідними даними
data_sets = [
    ("hello", "world"),
    ("programming", "python"),
    ("abcd", "bcdf")
]

for s1, s2 in data_sets:
    counter = CommonCharactersCounter(s1, s2)
    print(counter)
    print(f"Перший рядок: {s1}")
    print(f"Другий рядок: {s2}")
    print(f"Кількість однакових символів: {counter.common_count}")
    print("-" * 50)
```
