```python
# Cell 1: Імпорти та enum
from enum import Enum
import random

class BookStatus(Enum):
    AVAILABLE = "доступна"
    BORROWED = "взята"
    RESERVED = "зарезервована"
```

```python
# Cell 2: Класи
class Book:
    def __init__(self, title, author, genre):
        self.title = title
        self.author = author
        self.genre = genre
        self.status = BookStatus.AVAILABLE

class Student:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

class Statistics:
    def __init__(self):
        self.student_activity = {}
        self.book_popularity = {}
```

```python
# Cell 3: Клас Library
class Library:
    def __init__(self):
        self.books = []
        self.students = []
        self.stats = Statistics()
    
    def add_book(self, book):
        self.books.append(book)
    
    def add_student(self, student):
        self.students.append(student)
    
    def borrow_book(self, student, book):
        if book.status == BookStatus.AVAILABLE:
            book.status = BookStatus.BORROWED
            student.borrowed_books.append(book)
            self.stats.student_activity[student.name] = \
                self.stats.student_activity.get(student.name, 0) + 1
            self.stats.book_popularity[book.title] = \
                self.stats.book_popularity.get(book.title, 0) + 1
    
    def return_book(self, student, book):
        if book in student.borrowed_books:
            student.borrowed_books.remove(book)
            book.status = BookStatus.AVAILABLE
    
    def print_statistics(self):
        print("Активність студентів:")
        for student, count in self.stats.student_activity.items():
            print(f"{student}: {count} книг")
        print("\nПопулярність книг:")
        for book, count in self.stats.book_popularity.items():
            print(f"{book}: {count} разів")
```

```python
# Cell 4: Функція симуляції
def simulate_days(library, days=90):
    for _ in range(days):
        student = random.choice(library.students)
        book = random.choice(library.books)
        if book.status == BookStatus.AVAILABLE:
            library.borrow_book(student, book)
        else:
            library.return_book(student, book)
```

```python
# Cell 5: Виконання
library = Library()
library.add_book(Book("Python", "Гвідо", "Програмування"))
library.add_book(Book("Алгоритми", "Кормен", "ІТ"))
library.add_book(Book("Мережі", "Таненбаум", "ІТ"))
library.add_student(Student("Аня"))
library.add_student(Student("Богдан"))
library.add_student(Student("Іван"))
simulate_days(library)
library.print_statistics()
```

```python
#Методи класу
# Оголошуємо клас MyAnimals
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила один рік"

# Створюємо об’єкти
a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals("Собака", "Рекс", 2)

# Зберігаємо об’єкти у список для зручної обробки
animals_list = [a1, a2, a3]

# Застосовуємо методи speak() та add_one_year() до всіх об’єктів
for animal in animals_list:
    print(f"\nТваринка {animal.name} ({animal.animals}): {animal.speak()}")
    print(f"Вік до додавання року: {animal.age}")
    print(animal.add_one_year())
    print(f"Вік після додавання року: {animal.age}")
```

```python
#Змінні класу, глобальні змінні над всіма обєктами
class MyAnimals:
    # Змінні класу
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

        # Оновлюємо змінні класу
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1


# Вивід глобальних змінних перед створенням об'єктів
print(f"В мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}\n")

# Створюємо список даних для циклу
animal_data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5),
    ("Собака", "Рекс", 2)
]

# Створюємо об'єкти в циклі
objects = []
for data in animal_data:
    obj = MyAnimals(*data)
    objects.append(obj)

# Вивід глобальних змінних після створення об'єктів
print(f"\nВ мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")

# Доступ до змінної класу через об'єкт
a1 = objects[0]
print(f"Доступ через об'єкт a1: {a1.total_animal_types}")

# Перепризначення змінної класу для об'єкта (створюється атрибут об'єкта, клас не змінюється)
a1.total_animal_types = {"Собака"}
print(f"Після зміни у a1: {a1.total_animal_types} (атрибут об'єкта)")
print(f"З класу MyAnimals: {MyAnimals.total_animal_types} (залишається глобальна змінна класу)")
```

```python
#Статичні методи та змінна self
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


# Створюємо об’єкти
a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

# Викликаємо статичний метод для кожного об’єкта
print(f"Статичний метод для a1: {a1.call_pet()}")
print(f"Статичний метод для a2: {a2.call_pet()}")

# Також можна викликати безпосередньо через клас
print(f"Статичний метод через клас: {MyAnimals.call_pet()}")
```

```python
#Робота з properties
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    # Властивість для року, до якого тваринка буде з вами
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age
    
    # Нова властивість для віку в місяцях
    @property
    def age_in_months(self):
        return self.age * 12
    
    # Метод для додавання одного року
    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила один рік"


# Створюємо об'єкт
a1 = MyAnimals("Кіт", "Мурка", 5)

# Використовуємо властивості
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Вік тваринки в місяцях: {a1.age_in_months} місяців")

# Додаємо рік
a1.add_one_year()
print(f"Після додавання року, мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Вік тваринки в місяцях після додавання року: {a1.age_in_months} місяців")
```

```python
#Методи класу
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

    # Новий метод класу для перевірки першої літери name
    @classmethod
    def check_name_capital(cls, obj):
        if obj.name[0].isupper():
            return f"Ім'я '{obj.name}' задано правильно"
        else:
            return f"Ім'я '{obj.name}' має починатися з великої букви"


# Створюємо об’єкти
a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "rex", 2)  # тут маленька буква для прикладу

# Виводимо дані об'єктів
print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")

# Перевіряємо імена через метод класу
print(MyAnimals.check_name_capital(a1))
print(MyAnimals.check_name_capital(a2))

# Вивід загальної кількості
print(f"Загальна кількість об'єктів: {MyAnimals.total_objects}, а створених з CSV: {MyAnimals.total_csv}")
```

```python
#Магічні методи
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    # Офіційне представлення об'єкта
    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    # Зручне для користувача представлення
    def __str__(self) -> str:
        return f"Це об'єкт: {self.name} ({self.animals}), {self.age} рік(років)"

    # Додавання двох об'єктів (наприклад, сумуємо вік)
    def __add__(self, other):
        if isinstance(other, MyAnimals):
            return self.age + other.age
        return NotImplemented

    # Порівняння об'єктів за віком
    def __lt__(self, other):
        return self.age < other.age

    def __eq__(self, other):
        return self.age == other.age


# Створюємо об'єкти
a1 = MyAnimals("Собака", "Рекс", 2)
a2 = MyAnimals("Кіт", "Мурка", 5)

# Використання магічних методів
print(a1)           # викликає str
print(repr(a1))     # викликає repr
print(str(a2))      # викликає str

# Використовуємо add
print(f"Сумарний вік: {a1 + a2} років")

# Порівняння об'єктів (__lt__, __eq__)
print(f"Чи a1 молодший за a2? {a1 < a2}")
print(f"Чи a1 має такий же вік, як a2? {a1 == a2}")
```
